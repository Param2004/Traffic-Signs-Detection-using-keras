```python
import numpy as np
import cv2
from keras.models import load_model
from PIL import Image
```

```python
modal = load_model("model.h5")
```

```python
fontFace= cv2.FONT_HERSHEY_SIMPLEX
#Load the saved model

video = cv2.VideoCapture(0)

while True:
        _, frame = video.read()

        #Convert the captured frame into RGB
        im = Image.fromarray(frame, 'RGB')

        #Resizing into 128x128 because we trained the model with this image size.
        im = im.resize((32,32))
        img_array = np.array(im)
        img_array = img_array/255
        

        #Our keras model used a 4D tensor, (images x height x width x channel)
        #So changing dimension 128x128x3 into 1x128x128x3 
        img_array = np.expand_dims(img_array, axis=0)

        #Calling the predict method on model to predict 'me' on the image
        prediction = modal.predict(img_array)
        prediction = np.argmax(prediction, axis = 1)
        PV = np.amax(prediction)

        if   prediction == 0:  #print( 'Speed Limit 20 km/hr')
            cv2.putText(frame, str(prediction) , (120,35),fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
            cv2.putText(frame, str('Speed Limit 20 km/hr') , (180,75),fontFace, 0.75, (0,255,0), 2, cv2.LINE_AA)
        elif prediction == 1:  #print( 'Speed Limit 30 km/hr')
            cv2.putText(frame, str('Speed Limit 30 km/hr') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 2:  #print( 'Speed Limit 50 km/hr')
            cv2.putText(frame, str('Speed Limit 50 km/hr') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 3:  #print( 'Speed Limit 60 km/hr')
            cv2.putText(frame, str('Speed Limit 60 km/hr') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 4:  #print( 'Speed Limit 70 km/hr')
            cv2.putText(frame, str('Speed Limit 70 km/hr') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 5:  #print( 'Speed Limit 80 km/hr')
            cv2.putText(frame, str('Speed Limit 80 km/hr') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 6:  #print( 'End of Speed Limit 30 km/hr')
            cv2.putText(frame, str('End of Speed Limit 30 km/hr') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 7:  #print( 'Speed Limit 100 km/hr')
            cv2.putText(frame, str('Speed Limit 100 km/hr') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 8:  #print( 'Speed Limit 120 km/hr')
            cv2.putText(frame, str('Speed Limit 120 km/hr') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 9:  #print( 'No Passing')
            cv2.putText(frame, str('No Passing') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 10: #print( 'No Passing for vehicles over 3.5 metric tons')
            cv2.putText(frame, str( 'No Passing for vehicles over 3.5 metric tons') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 11: #print( 'Right-of-way of next intersection')
            cv2.putText(frame, str('Right-of-way of next intersection') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 12: #print( 'Priority Road')
            cv2.putText(frame, str('Priority Road') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 13: #print( 'Yield')
            cv2.putText(frame, str('Yield') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 14: #print( 'Stop')
            cv2.putText(frame, str('Stop') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 15: #print( 'NO vehicles')
            cv2.putText(frame, str( 'NO vehicles') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 16: #print( 'vehicles over 3.5 tons are probhited')
            cv2.putText(frame, str('vehicles over 3.5 tons are probhited') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 17: #print( 'No Entry')
            cv2.putText(frame, str('No Entry') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 18: #print( 'General Caution')
            cv2.putText(frame, str('General Caution') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 19: #print( 'Dangerous curve to the left')
            cv2.putText(frame, str('Dangerous curve to the left') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 20: #print( 'Dangerous curve to the right')
            cv2.putText(frame, str('Dangerous curve to the right') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 21: #print( 'Dangerous curve')
            cv2.putText(frame, str( 'Dangerous curve') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 22: #print( 'Bumpy Road')
            cv2.putText(frame, str('Bumpy Road') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 23: #print( 'Slippery Road')
            cv2.putText(frame, str('Slippery Road') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 24: #print( 'Road narrows on the right')
            cv2.putText(frame, str('Road narrows on the right') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 25: #print( 'Road work')
            cv2.putText(frame, str( 'Road work') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 26: #print( 'Traffic signals')
            cv2.putText(frame, str('Traffic signals') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 27: #print( 'Pedestrians')
            cv2.putText(frame, str('Pedestrians') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 28: #print( 'Children Crossing')
            cv2.putText(frame, str('Children Crossing') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 29: #print( 'Bicycles Crossing')
            cv2.putText(frame, str('Bicycles Crossing') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 30: #print( 'Beware of ice/snow')
            cv2.putText(frame, str('Beware of ice/snow') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 31: #print( 'Wild animals crossing')
            cv2.putText(frame, str('Wild animals crossing') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 32: #print( 'End of all speed and passing limits')
            cv2.putText(frame, str('End of all speed and passing limits') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 33: #print( 'Turn right ahead')
            cv2.putText(frame, str('Turn right ahead') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 34: #print( 'Turn left ahead')
            cv2.putText(frame, str('Turn left ahead') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 35: #print( 'Ahead only')
            cv2.putText(frame, str('Ahead only') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 36: #print( 'Go straight or right')
            cv2.putText(frame, str('Go straight or right') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 37: #print( 'Go straight or left')
            cv2.putText(frame, str('Go straight or left') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 38: #print( 'Keep right')
            cv2.putText(frame, str('Keep right') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 39: #print( 'Keep left')
            cv2.putText(frame, str('Keep left') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 40: #print( 'Roundabout Mandatory')
            cv2.putText(frame, str('Roundabout Mandatory') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 41: #print( 'End of no pasiing')
            cv2.putText(frame, str('End of no pasiing') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
        elif prediction == 42: #print( 'End of no passing by vehicles over 3.5 metric tons')
            cv2.putText(frame, str('End of no passing by vehicles over 3.5 metric tons') , (20,35) ,fontFace, 0.75, (0,0,255), 2, cv2.LINE_AA)
            
        cv2.imshow("Capturing", frame)
        key=cv2.waitKey(1)
        if key == ord('q'):
                break
video.release()
cv2.destroyAllWindows()
```
